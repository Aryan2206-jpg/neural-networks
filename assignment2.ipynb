What are epoch, batch size, and Stratified ? Please explain them.
(a) Epoch
An epoch refers to one complete pass of the entire training dataset through the model.

Explanation
When the model sees all training samples once, it completes 1 epoch
Training usually requires multiple epochs for better learning
Example
If dataset = 1000 samples

1 epoch = model processes all 1000 samples once
More epochs → better learning (but too many can cause overfitting)

(b) Batch Size
Batch size is the number of training samples processed before updating the model weights.

Explanation
Instead of processing the whole dataset at once, data is divided into batches
After each batch, weights are updated
Example
Dataset = 1000 samples Batch size = 100

Number of batches = 10 per epoch

Types
Batch Gradient Descent → uses entire dataset
Mini-batch Gradient Descent → uses small batches (most common)
Stochastic Gradient Descent (SGD) → batch size = 1
(c) Stratified (Sampling / Splitting)
Stratified means maintaining the same proportion of classes in training and testing data.

Why Important?
Prevents class imbalance
Ensures fair representation
Example
If dataset:

80% Pass
20% Fail
Then after split:

Train and test sets should also have 80:20 ratio
Commonly used in classification problems

2) Explain the different types for activation functions?
Activation functions introduce non-linearity into neural networks.

(1) Sigmoid Function
[ \sigma(x) = \frac{1}{1 + e^{-x}} ]

Output: (0,1)
Used in binary classification
Advantages:

Smooth curve
Probabilistic output
Disadvantages:

Vanishing gradient problem
(2) Tanh Function
[ tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} ]

Output: (-1, 1)
Better than sigmoid
(3) ReLU (Rectified Linear Unit)
[ f(x) = \max(0, x) ]

Most widely used
Fast computation
Issue: Dead neuron problem

(4) Leaky ReLU
[ f(x) =
{x,x>0 0.01x,x≤0
]

Fixes ReLU problem
Allows small negative values
(5) Softmax Function
[ P(y_i) = \frac{e^{z_i}}{\sum e^{z_j}} ]

Used in multi-class classification
Output = probabilities summing to 1
Summary Table
Function	Range	Use
Sigmoid	(0,1)	Binary classification
Tanh	(-1,1)	Hidden layers
ReLU	[0,∞)	Deep learning
Leaky ReLU	(-∞,∞)	Fix ReLU issues
Softmax	(0,1)	Multi-class

4) What is a layer? Define ANN and CNN layers?
What is a Layer?
A layer is a collection of neurons that processes input data and passes output to the next layer.

Layers in ANN
1. Input Layer
Receives input features
2. Hidden Layers
Perform computations
Apply weights + activation functions
3. Output Layer
Produces final result
Example ANN Flow
Input → Hidden Layer → Output

Layers in CNN
1. Convolution Layer
Applies filters (kernels)
Extracts features
2. Activation Layer
Applies ReLU or other functions
3. Pooling Layer
Reduces dimensions
Example: Max pooling
4. Fully Connected Layer
Final classification
CNN Flow
Image → Convolution → Pooling → Fully Connected → Output
