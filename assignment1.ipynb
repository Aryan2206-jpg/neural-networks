# **Explain the role of weights in a neuron**.
In a neural network, a neuron is the basic computational unit that receives multiple inputs and produces an output.

# **How a Neuron Works**
A neuron performs two main operations:

Multiply each input with a corresponding weight
Add all weighted inputs and pass through an activation function
Mathematically: [ Z = w1x1 + w2x2 + w3x3 + ... + b ]

Where:

(x) = inputs
(w) = weights
(b) = bias
(Z) = weighted sum
# **What are Weights?**
Weights are learnable parameters that control how much influence each input has on the output.

Each input has its own weight

Weights can be:

Positive → increase influence
Negative → decrease influence
During training, weights are updated using optimization algorithms

Why are Weights Important?
They allow the model to learn patterns from data
They determine the decision boundary in classification problems
Proper weight adjustment leads to accurate predictions
Example
Suppose:

Study hours = 5
Weight = 0.8
Then contribution = (5 × 0.8 = 4)

This shows how strongly "study hours" affects the final prediction.

# 2) **What is an activation function?**
An activation function decides whether a neuron should be activated or not.

After calculating the weighted sum, the result is passed through this function.

# **Why Do We Need Activation Functions?**
Without activation functions:

The neural network becomes linear
Cannot learn complex patterns
Activation functions introduce non-linearity, making deep learning powerful.

# **Common Activation Functions**
1. Sigmoid Function
Output range: (0, 1)
Used in binary classification
2. ReLU (Rectified Linear Unit)
Very fast and widely used
Converts negative values to 0
3. Tanh Function
Output range: (-1, 1)
Better than sigmoid in some cases
3) Define probability distribution in ML context.
A probability distribution describes how likely different outcomes are.

# **Definition**
It is a function that assigns probabilities to all possible values of a random variable.

# **Types of Probability Distributions**
1. Discrete Distribution
Deals with countable values
Example: Number of students passing
Example: Bernoulli Distribution

Only two outcomes: 0 or 1
2. Continuous Distribution
Deals with continuous values
Example: Height, weight
Example: Normal Distribution

Bell-shaped curve
Mean = center
Symmetrical
Why Important in ML?
Helps in modeling uncertainty

Used in:

Naive Bayes classifier
Gaussian models
Prediction probabilities
Example
If a model predicts:

0.9 → high probability of passing
0.1 → low probability
These values come from a probability distribution.

4) What is gradient in optimization?
The gradient is a fundamental concept in optimization used to train ML models.

# **Definition**
The gradient is a vector that shows:

Direction of steepest increase
Rate of change of a function
Mathematical Idea
If loss function = (L(w))

Then gradient: [ \nabla L = \frac{\partial L}{\partial w} ]

Role in Machine Learning
Used in Gradient Descent
Helps update weights to minimize error
Gradient Descent Update Rule
[ w = w - \eta \cdot \nabla L ]

Where:

(w) = weight
(\eta) = learning rate
(\nabla L) = gradient
Intuition
Gradient tells: “Which direction should I move to reduce error?”
Example
If error is increasing:

Move in opposite direction of gradient
Types of Gradient Descent
Batch Gradient Descent
Stochastic Gradient Descent (SGD)
Mini-batch Gradient Descent
